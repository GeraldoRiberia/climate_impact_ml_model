```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("aqi_data.csv")
df.head()
```

```python
cnt = df['Country'].unique()
print(len(cnt))
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(by=["Country", "Date"]).reset_index(drop=True)
df.head()
```

```python
df[df["Country"] == "India"]
```

```python
df.isna().sum()
```

```python
df.rename(columns={"AQI Value":"AQI"}, inplace=True)
```

```python
data = df[df['Country']=='India']
data
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.Date, data['AQI'], color='red', label='avgtemp')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('AQI Trends')
plt.legend()
plt.show()
```
